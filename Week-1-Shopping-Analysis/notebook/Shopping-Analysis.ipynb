## Data Cleaning

```python
df["final_price"].head()
```

```python
df["final_price"] = df["final_price"].str.replace('"', '', regex=False)
df["final_price"] = df["final_price"].str.replace('₹', '', regex=False)
df["final_price"] = df["final_price"].str.replace(',', '', regex=False)

df["final_price"] = pd.to_numeric(df["final_price"], errors="coerce")

df["final_price"].head()
```

```python
df["final_price"].dtype
```

### Observation

The final_price column had text in it like currency symbols and commas so it couldn't be used for calculations directly

```python
df["discount"] = df["discount"].fillna(0)
```

```python
df["discount"].isnull().sum()
```

```python
df.info()
```

### observation

I removed the ₹ symbol and commas from final_price and changed it to a number (float)

I also filled the missing values in the discount column

## Feature Engineering

```python
df["price_difference"] = df["initial_price"] - df["final_price"]
```

```python
df["popularity"] = df["rating"] * df["ratings_count"]
```

```python
df.head()
```

### observation

I made two new columns
price_difference shows how much the price dropped from the original price to the final price
popularity is calculated by multiplying rating with ratings_count so it shows which products are both liked and reviewed a lot

```python
df["price_difference"].describe()
```

### observation
The price_difference column represents the amount saved by customers after discounts

Products with larger price differences provide greater savings while products with smaller differences offer little or no discount

## Basic Data Analysis

### Bivariate Analysis

```python
plt.figure(figsize=(6,4))
plt.scatter(df["rating"], df["ratings_count"])
plt.xlabel("Rating")
plt.ylabel("Ratings Count")
plt.title("Rating vs Ratings Count")
plt.show()
```

## observation 
This plot shows rating compared to number of ratings

Some products have both a high rating and a lot of reviews but not all high-rated products get a lot of reviews

So a product's popularity depends on both how good it is and how many people have actually bought/reviewed it

# Data Visualization

```python
plt.figure(figsize=(8,5))
plt.hist(df["final_price"])
plt.title("Distribution of Final Price")
plt.xlabel("Final Price")
plt.ylabel("Count")
plt.show()
```

### observation
Most products are priced in the low to medium range

```python
top_products = df.nlargest(10,"popularity")

plt.figure(figsize=(10,5))
plt.bar(top_products["title"], top_products["popularity"])
plt.xticks(rotation=90)
plt.title("Top 10 Popular Products")
plt.show()
```

### observation
SASSAFRAS and Athena have the highest popularity scores much higher than other products

This means a lot of people have rated these products and given them good ratings

```python
plt.figure(figsize=(8,5))
sns.boxplot(x=df["final_price"])
plt.title("Final Price Distribution")
plt.show()
```

### Observation

A few products are priced much higher than the rest so they show up as outliers in the boxplot
